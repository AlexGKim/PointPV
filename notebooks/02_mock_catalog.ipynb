```python
import numpy as np
import matplotlib.pyplot as plt
import sys
from pathlib import Path
sys.path.insert(0, str(Path('..').resolve()))
```

```python
# Synthetic catalog (no abacusutils needed)
sys.path.insert(0, str(Path('../scripts').resolve()))
from generate_mock import generate_synthetic_catalog

catalog = generate_synthetic_catalog(n=500, seed=42)
print('Keys:', list(catalog.keys()))
print('N =', len(catalog['ra']))
```

```python
from pointpv.mock.catalog import eta_to_velocity

u = eta_to_velocity(catalog['eta'], catalog['z_obs'])
print(f'u: mean={u.mean():.1f}  std={u.std():.1f} km/s')

plt.figure(figsize=(6, 4))
plt.hist(u, bins=40)
plt.xlabel('u = v_los [km/s]')
plt.title('Velocity distribution (noise-only mock)')
plt.tight_layout()
plt.show()
```
