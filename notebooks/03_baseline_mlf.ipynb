# 03 — Baseline MLF Likelihood

Evaluate the MLF (Cholesky) log-likelihood and scan over fsigma8.

```python
import numpy as np
import matplotlib.pyplot as plt
import sys
from pathlib import Path
sys.path.insert(0, str(Path('..').resolve()))
sys.path.insert(0, str(Path('../scripts').resolve()))
```

```python
from generate_mock import generate_synthetic_catalog
from pointpv.mock.catalog import eta_to_velocity
from pointpv.likelihood.mlf import scan_fsigma8

catalog = generate_synthetic_catalog(n=200, seed=42)
u = eta_to_velocity(catalog['eta'], catalog['z_obs'])

fsigma8_grid = np.linspace(0.2, 0.8, 20)
results = scan_fsigma8(u, catalog, fsigma8_values=fsigma8_grid, verbose=True)
```

```python
logL = results['logL']
fs8 = results['fsigma8']

plt.figure(figsize=(7, 4))
plt.plot(fs8, logL - logL.max())
plt.axhline(-0.5, color='gray', linestyle='--', label=r'$-\frac{1}{2}\sigma$')
plt.xlabel(r'$f\sigma_8$')
plt.ylabel(r'$\Delta \log L$')
plt.title('MLF: log-likelihood profile')
plt.legend()
plt.tight_layout()
plt.show()

best = fs8[np.argmax(logL)]
print(f'Best-fit fsigma8 = {best:.3f}')
```
