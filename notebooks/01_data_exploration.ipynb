# 01 — Data Exploration

Explore the AbacusSummit light cone mock catalog: sky distribution, redshift histogram, velocity distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import sys
sys.path.insert(0, str(Path('..').resolve()))
```

```python
# Load catalog (generate first if needed):
# !python scripts/generate_mock.py --n 1000 --synthetic
from pointpv.mock.catalog import load_catalog
catalog = load_catalog('../data/mock_1000.npz')
print({k: v.shape for k, v in catalog.items()})
```

```python
fig, axes = plt.subplots(1, 3, figsize=(14, 4))

axes[0].scatter(catalog['ra'], catalog['dec'], s=1, alpha=0.5)
axes[0].set_xlabel('RA [deg]')
axes[0].set_ylabel('Dec [deg]')
axes[0].set_title('Sky Distribution')

axes[1].hist(catalog['z_obs'], bins=40)
axes[1].set_xlabel('z_obs')
axes[1].set_title('Redshift Distribution')

axes[2].hist(catalog['v_r_true'], bins=40)
axes[2].set_xlabel('v_r_true [km/s]')
axes[2].set_title('True Peculiar Velocity')

fig.tight_layout()
plt.show()
```
